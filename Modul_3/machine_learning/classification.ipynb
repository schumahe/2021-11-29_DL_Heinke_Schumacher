### k-Nearest-Neighbor-Klassifikator

```python
import matplotlib.pyplot as plt
%matplotlib inline
fig, axes = plt.subplots(3, 10, figsize=(10, 5))
for ax, img in zip(axes.ravel(), digits.images):
    ax.imshow(img, cmap=plt.cm.gray_r)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    digits['data'], digits['target'], random_state=1)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_clf = KNeighborsClassifier(n_neighbors=1)
```

```python
knn_clf.fit(X_train, y_train)
```

```python
knn_clf.predict(X_test)

# Der Testdatensatz sind neue Werte, auf die das Modell angewendet werden kann
```

```python
knn_clf.score(X_test, y_test)
```

```python
knn_clf = KNeighborsClassifier(n_neighbors=3)
```

```python
knn_clf.fit(X_train, y_train)
```

```python
knn_clf.predict(X_test)
```

```python
knn_clf.score(X_test, y_test)
```

### Random-Forest-Klassifikator

```python
from sklearn.ensemble import RandomForestClassifier
random_forest_cfl = RandomForestClassifier(random_state=1)
```

```python
random_forest_cfl.fit(X_train, y_train)
```

```python
random_forest_cfl.score(X_test, y_test)
```

### Multi-Layer-Perception, künstliches, neurales Netz (Multi-Layer-Perceptron) mit einem Hidden-Layer mit 100 Nodes.

```python
from sklearn.neural_network import MLPClassifier
```

```python
mlpc = MLPClassifier(random_state=1)
mlpc.fit(X_train, y_train)
```

```python
mlpc.score(X_test, y_test)
```

```python
MLPClassifier(random_state=1, hidden_layer_sizes=(200, 100, 20)).fit(
    X_train, y_train).score(X_test, y_test) 

# die Anzahl an Hidden-Layer und Anzahl an Nodes hier erhöht auf 3 Schichten mit mit 200, 100 und 20 Nodes
```
