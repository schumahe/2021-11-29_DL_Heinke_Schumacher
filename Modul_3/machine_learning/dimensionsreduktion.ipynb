## Machine Learning: DIMENSIONSREDUKTION
***

```python
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
from sklearn.decomposition import PCA
pca = PCA(random_state=1, n_components=2)

# Wir wollen mit einer Pricinple Component Analysis (PCA) die Dimensionreduktion durchführen. 
# Hierzu laden wir zuerst die Klasse und erzeugen eine Instanz davon. Wir geben an, 
# dass wir danach nur zwei Dimensionen (“components”) erhalten möchten. 
# Dadurch lassen sie die Daten später auf in einem Plot mit zwei Achsen abbilden.
```

```python
pca_result = pca.fit_transform(digits.data)
```

```python
pca_result.shape
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
fig, ax = plt.subplots()
plt.scatter(pca_result[:, 0], pca_result[:, 1], c=digits.target, cmap="Set1")
plt.colorbar()
```

```python
from sklearn.manifold import TSNE
tsne = TSNE(random_state=1, n_components=2)
tsne_result = tsne.fit_transform(digits.data)
```

```python
fig, ax = plt.subplots()
plt.scatter(tsne_result[:, 0], tsne_result[:, 1], c=digits.target, cmap="Set1")
plt.colorbar()
```
